```python
numeric_columns=['Model_Score','Balanced Accuracy', 'Precision', 'Recall',\
                 'f1-score', 'Confusion TP', 'Confusion FP', 'Confusion FN', 'Confusion TN']
```

```python
for feature in numeric_columns:
    print(f'\n\033[33m\033[1m{feature}\033[0m by Balanced/Unbalanced Ratio')
    for ratio in results_df['Proportion'].unique():
        prec_max=results_df[feature].loc[results_df['Proportion']==ratio].max()
        prec_mean=results_df[feature].loc[results_df['Proportion']==ratio].mean()
        print(f"{ratio} \tMax : {prec_max:.4f}\tMean : {prec_mean:.4f}", end='')
        print(f"\tMin : {results_df[feature].loc[results_df['Proportion']==ratio].min():.4f}")
```

```python
print(f'\n\033[33m\033[1mPrecision\033[0m by Balanced/Unbalanced Ratio')
for ratio in results_df['Proportion'].unique():
    prec_max=results_df['Precision'].loc[results_df['Proportion']==ratio].max()
    prec_mean=results_df['Precision'].loc[results_df['Proportion']==ratio].mean()
    print(f"{ratio} \tMax : {prec_max:.4f}\tMean : {prec_mean:.4f}", end='')
    print(f"\tMin : {results_df['Precision'].loc[results_df['Proportion']==ratio].min():.4f}")

print(f'\n\033[33m\033[1mRecall\033[0m by Balanced/Unbalanced Ratio')
for ratio in results_df['Proportion'].unique():
    print(f"{ratio} \tMax : {results_df['Recall'].loc[results_df['Proportion']==ratio].max():.4f}",end='')
    print(f"\tMean : {results_df['Recall'].loc[results_df['Proportion']==ratio].mean():.4f}",end='')
    print(f"\tMin : {results_df['Recall'].loc[results_df['Proportion']==ratio].min():.4f}")

print(f'\n\033[33m\033[1mf1-scores\033[0m by Balanced/Unbalanced Ratio')
for ratio in results_df['Proportion'].unique():
    print(f"{ratio} \tMax : {results_df['f1-score'].loc[results_df['Proportion']==ratio].max():.4f} ",end='')
    print(f"\tMean : {results_df['f1-score'].loc[results_df['Proportion']==ratio].mean():.4f}",end='')
    print(f"\tMin : {results_df['f1-score'].loc[results_df['Proportion']==ratio].min():.4f}")
```
