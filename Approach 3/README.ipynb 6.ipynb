# Stroke Prediction Model

## Project Overview

This project aims to develop a machine learning model to predict the occurrence of strokes in individuals based on various health indicators. We utilized several classification algorithms, addressing the imbalance in the dataset with Synthetic Minority Over-sampling Technique (SMOTE) and refining our models with feature engineering and hyperparameter tuning.
